# Checker's Game

```python
import numpy as np
import random
import copy
```

### Initialize the board to blank characters

```python
bs = 8
def initializeBoard():
    board  = [[' ','B',' ','B',' ','B',' ','B'],
              ['B',' ','B',' ','B',' ','B',' '],
              [' ','B',' ','B',' ','B',' ','B'],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              [' ',' ',' ',' ',' ',' ',' ',' '],
              ['W',' ','W',' ','W',' ','W',' '],
              [' ','W',' ','W',' ','W',' ','W'],
              ['W',' ','W',' ','W',' ','W',' ']]
    return board
```

### Checks if game has reached its end

```python
def isGameOver(board, symbol):
    sym = ''
    if symbol == 'W' or symbol == 'KW':
        sym = 'B'
    else:
        sym = 'W'
    for i in range(bs):
        for j in range(bs):
            if(board[i][j]==sym or board[i][j]=='K'+sym):
                return 0
    return 1
```

### Move 2 steps

```python
def checkCross(board,symbol,r1,c1,r2,c2,r3,c3):
    if(r3 < 0 or r3 >= bs or c3 < 0 or c3 >= bs):
        return False
    if(board[r3][c3]!=' '):
        return False
    if (symbol == 'W'):
        if (board[r1][c1] == 'W' and r3 > r1):
            return False
        if (board[r2][c2] != 'B' and board[r2][c2] != 'KB'):
            return False
        return True;
    
    else:
        if (board[r1][c1] == 'B' and r3 < r1):
            return False
        if (board[r2][c2] != 'W' and board[r2][c2] != 'KW'):
            return False
        return True
```

### Move 1 step

```python
def checkMove(board,symbol,r1,c1,r2,c2):
    if (r2 < 0 or r2 >= bs or c2 < 0 or c2 >= bs):
        return False

    if (board[r2][c2] != ' '):
        return False

    if (symbol == 'W'):
        if (board[r1][c1] == 'W' and r2 > r1):
            return False
        return True
    
    else:
        if (board[r1][c1] == 'B' and r2 < r1):
            return False
        return True
```

```python
def makeKings(board):
    for j in range(bs):
        if(board[0][j]=='W'):
            board[0][j]='KW'
        if(board[bs-1][j]=='B'):
            board[bs-1][j]='KB'
```

### All legal next moves

```python
def getLegalMoves(boardState, symbol):
    
    legalMoves = []
    
    if(symbol=='B'):
        playerKing='KB'
    else:
        playerKing='KW'
    for i in range(len(boardState[0])):
        for j in range(len(boardState[0])):
            if(boardState[i][j]==symbol or boardState[i][j]==playerKing):
                if(checkCross(boardState,symbol,i,j,i+1,j+1,i+2,j+2)):
                    tempBoard = copy.deepcopy(boardState)
                    tempBoard[i+2][j+2]=tempBoard[i][j]
                    tempBoard[i][j]=' '
                    tempBoard[i+1][j+1]=' '
                    makeKings(tempBoard)
                    legalMoves.append(tempBoard)
                if(checkCross(boardState,symbol,i,j,i-1,j+1,i-2,j+2)):
                    tempBoard = copy.deepcopy(boardState)
                    tempBoard[i-2][j+2]=tempBoard[i][j]
                    tempBoard[i][j]=' '
                    tempBoard[i-1][j+1]=' '
                    makeKings(tempBoard)
                    legalMoves.append(tempBoard)
                if(checkCross(boardState,symbol,i,j,i+1,j-1,i+2,j-2)):
                    tempBoard = copy.deepcopy(boardState)
                    tempBoard[i+2][j-2]=tempBoard[i][j]
                    tempBoard[i][j]=' '
                    tempBoard[i+1][j-1]=' '
                    makeKings(tempBoard)
                    legalMoves.append(tempBoard)
                if(checkCross(boardState,symbol,i,j,i-1,j-1,i-2,j-2)):
                    tempBoard = copy.deepcopy(boardState)
                    tempBoard[i-2][j-2]=tempBoard[i][j]
                    tempBoard[i][j]=' '
                    tempBoard[i-1][j-1]=' '
                    makeKings(tempBoard)
                    legalMoves.append(tempBoard)
    if(len(legalMoves)==0):
        for i in range(len(boardState[0])):
            for j in range(len(boardState[0])):
                if(boardState[i][j]==symbol or boardState[i][j]==playerKing):
                    if(checkMove(boardState,symbol,i,j,i+1,j+1)):
                        tempBoard = copy.deepcopy(boardState)
                        tempBoard[i+1][j+1]=tempBoard[i][j]
                        tempBoard[i][j]=' '
                        makeKings(tempBoard)
                        legalMoves.append(tempBoard)
                    if(checkMove(boardState,symbol,i,j,i-1,j+1)):
                        tempBoard = copy.deepcopy(boardState)
                        tempBoard[i-1][j+1]=tempBoard[i][j]
                        tempBoard[i][j]=' '
                        makeKings(tempBoard)
                        legalMoves.append(tempBoard)
                    if(checkMove(boardState,symbol,i,j,i+1,j-1)):
                        tempBoard = copy.deepcopy(boardState)
                        tempBoard[i+1][j-1]=tempBoard[i][j]
                        tempBoard[i][j]=' '
                        makeKings(tempBoard)
                        legalMoves.append(tempBoard)
                    if(checkMove(boardState,symbol,i,j,i-1,j-1)):
                        tempBoard = copy.deepcopy(boardState)
                        tempBoard[i-1][j-1]=tempBoard[i][j]
                        tempBoard[i][j]=' '
                        makeKings(tempBoard)
                        legalMoves.append(tempBoard)
    
    return legalMoves
```

### Get Features of the Checker's Learning Problem

```python
def getFeatures(board, symbol1, symbol2):
    x = np.zeros(7)
    x[0]=1
    for i in range(bs):
        for j in range(bs):
            if(board[i][j]==symbol1):
                x[1]+=1
            elif(board[i][j]==symbol2):
                x[2]+=1
            elif(board[i][j]==('K'+symbol1)):
                x[3]+=1
            elif(board[i][j]==('K'+symbol2)):
                x[4]+=1
            if(board[i][j]==symbol1 or board[i][j]==('K'+symbol1)):
                if(checkCross(board,symbol1,i,j,i+1,j+1,i+2,j+2)):
                    x[5]+=1
                if(checkCross(board,symbol1,i,j,i-1,j+1,i-2,j+2)):
                    x[5]+=1
                if(checkCross(board,symbol1,i,j,i+1,j-1,i+2,j-2)):
                    x[5]+=1
                if(checkCross(board,symbol1,i,j,i-1,j-1,i-2,j-2)):
                    x[5]+=1
            elif(board[i][j]==symbol2 or board[i][j]==('K'+symbol2)):
                if(checkCross(board,symbol2,i,j,i+1,j+1,i+2,j+2)):
                    x[6]+=1
                if(checkCross(board,symbol2,i,j,i-1,j+1,i-2,j+2)):
                    x[6]+=1
                if(checkCross(board,symbol2,i,j,i+1,j-1,i+2,j-2)):
                    x[6]+=1
                if(checkCross(board,symbol2,i,j,i-1,j-1,i-2,j-2)):
                    x[6]+=1
    return x
```

### Weighted Sum score of the Intermediate Board States

```python
def getIntermediateBoardScore(weights, features):
    weights = np.array(weights).reshape((len(weights),1))
    features = np.array(features).reshape((len(features),1))
    boardScore = np.dot(weights.T,features)
    return(boardScore[0][0])
```

### Choose Optimal next move

```python
def chooseMove(board,symbol1,symbol2,weights):
    legalMoves = getLegalMoves(board,symbol1)
    if len(legalMoves)==0:
        return -1
    legalMoveScores = [getIntermediateBoardScore(weights, getFeatures(i,symbol1,symbol2)) for i in legalMoves]
    return legalMoves[np.argmax(legalMoveScores)]
```

### Choose random move for opponent during the training of the model

```python
def chooseRandomMove(board,symbol):
    legalMoves = getLegalMoves(board,symbol)
    if(len(legalMoves)==0):
        return -1
    return random.choice(legalMoves)
```

### Get Checker Game history when model plays against the random model chooser

```python
def getCheckerHistory(symbols, w1, board):
    checkerHistory = []
    gameStatusFlag = True
    
    tempBoard = copy.deepcopy(board)
    while(gameStatusFlag):
        tempBoard = chooseMove(tempBoard,symbols[0],symbols[1],w1)
        if(tempBoard==-1):
            break
        checkerHistory.append(tempBoard)
        gameStatusFlag = not isGameOver(tempBoard,symbols[0])
        if(gameStatusFlag == False):
            break
        tempBoard = chooseRandomMove(tempBoard,symbols[1])
        if(tempBoard==-1):
            break
        checkerHistory.append(tempBoard)
        gameStatusFlag =  not isGameOver(tempBoard,symbols[1])
    return checkerHistory
```

### Final Board Score

```python
def getFinalBoardScore(board, symbol1, symbol2):
    # Draw
    score = 0
    # Player wins
    if isGameOver(board,symbol1):
        score = 100
    # Opponent wins
    elif isGameOver(board,symbol2):
        score = -100   
    return score
```

### Finds scores of all board states in the game history and returns the trained data

```python
def getTrainingData(symbol1,symbol2,checkerHistory):
    trainingData=[]
    for i in range(len(checkerHistory)-1):
        features = getFeatures(checkerHistory[i+1],symbol1,symbol2)
        trainingData.append([features,getFinalBoardScore(checkerHistory[-1],symbol1,symbol2)])
    trainingData.append([getFeatures(checkerHistory[-1],symbol1,symbol2),
        getFinalBoardScore(checkerHistory[-1],symbol1,symbol2)])
    return trainingData
```

### Least Mean Square (LMS) weight update rule

```python
def gradientDescent(weights, trainData, lr=0.005):
    error = 0
    for x,y in trainData:
        target = y
        output = getIntermediateBoardScore(weights, x)
        weights = weights + (lr * (target - output) * np.array(x))
        error += 0.5*(target - output)**2
    return weights
```

```python
symbols = ('W','B')
numOfTrainingExamples = 100
```

### Game Status Count

```python
def getGameResult(token1, token2, gameStatusCount, gameHistory):

    finalScore = getFinalBoardScore(gameHistory[-1], token1, token2)
    
    if(finalScore == 100):
        gameStatusCount[0] += 1
    elif(finalScore == -100):
        gameStatusCount[1] += 1
    else:
        gameStatusCount[2] += 1
    
    return gameStatusCount
```

### Training the Model

The **Experiment Generator** gives the initial board state.

The **Performance System** solves the task, here it is playing a game using learnt weights.

The **Critic** takes input the game history and outputs set of training samples of the target function.

The **Generalizer** here is the LMS algorithm that takes input the generated training samples and updates the weights.

```python
def displayBoard(board):
    for i in range(bs):
        for j in range(bs):
            print(board[i][j],end=' | ')
        print()
        for _ in range(4*bs):
            print("~",end='')
        print()
```

```python
weights = [np.random.rand(7)]
gameStatusCount = [0,0,0]

for i in range(numOfTrainingExamples):
    
    # Experiment Generator
    initialBoardState = initializeBoard()

    # Performance System
    # board hisory of player 1 [optimal move based on learnt weights(weights[0])] and player 2 [random move]
    checkerHistory = getCheckerHistory(symbols,weights[0],initialBoardState)

    # Critic
    # for each game in board history, create features, find score
    trainData1 = getTrainingData(symbols[0],symbols[1],checkerHistory)
    # game final result
    gameStatusCount = getGameResult(symbols[0], symbols[1], gameStatusCount, checkerHistory)

    # Generalizer
    weights = [gradientDescent(weights[0],trainData1)]

# Weights
Weight =  weights[0]
print("Weights: \n", Weight)
print("\n Training Results: ")
print(f"\n P1 Wins: {gameStatusCount[0]}")
print(f"\n P2 Wins: {gameStatusCount[1]}")
print(f"\n Draws: {gameStatusCount[2]}")
```
